###Calculating Sentiment Score

```python
data = (dataset['TWEET'])
score_list = []
for tweet in data:
  print(tweet)
  final_score = 0
  added_score = 0
  count_of_sentiment_words = 0
  nltk_tokens = nltk.word_tokenize(tweet)
 
  for word in nltk_tokens:
  
    if word in afinn_dictionary:
      count_of_sentiment_words =  count_of_sentiment_words +1
      score_of_word = afinn_dictionary[word]
      print(word, score_of_word)
      added_score = added_score + score_of_word
    elif word in hindi_dictionary:
      count_of_sentiment_words =  count_of_sentiment_words +1
      score_of_word = hindi_dictionary[word]
      print(word, score_of_word)
      added_score = added_score + score_of_word
  
    if count_of_sentiment_words!=0:
      final_score = added_score/count_of_sentiment_words
    else:
      final_score = 0
  print(final_score)
  score_list.append(final_score)
```

```python
print(score_list)
```

###Classifying Tweets Using Sentiment Labels

```python
sentiment_label = []
for score in score_list:
  if score > 0:
    sentiment_label.append('positive')
  elif score < 0:
    sentiment_label.append('negative')
  else:
    sentiment_label.append('neutral')
print(sentiment_label)
```
