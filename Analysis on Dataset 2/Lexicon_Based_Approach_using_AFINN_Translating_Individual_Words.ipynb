Function for Calculating Sentiment Score

```python
def sentiment_score_calc(list_of_eng):
  final_score = 0
  added_score = 0
  count_of_sentiment_words = 0
  for word in list_of_eng:
    #print(word)
      if word in mapped_words:
        count_of_sentiment_words =  count_of_sentiment_words +1
      #print(count_of_sentiment_words)
        score_of_word = mapped_words[word]
       # print(word, score_of_word)
        added_score = added_score + score_of_word
  if count_of_sentiment_words!=0:
      final_score = added_score/count_of_sentiment_words
  else:
      final_score = 0
  return final_score
```

Tokenizing the Tweets and Translating Individual Words 

```python
dataset = pd.read_csv('/content/Book1.csv',error_bad_lines=False)
dataset.TWEET=dataset.TWEET.astype(str)
dataset['TWEET'] = dataset['TWEET'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopwords_list)]))

data = (dataset['TWEET'])
score_list=[]
list_of_converted_sentences =[]
list_of_all_tagged_words=[]

for text in data:
  score = 0
  list_of_english = []
  text= text.lower()
  print(text)
  punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~0123456789+'''
  no_punct = ""
  for char in text:
      if char not in punctuations:
          no_punct = no_punct + char
  #print(no_punct)
  nltk_tokens = nltk.word_tokenize(no_punct)
  #print(nltk_tokens)
  list_of_tagged_words=[]
  list_of_english_words=[]
  list_of_romanized_hindi_words=[]  
  list_of_devanagri_words=[]
  list_of_translated_words=[]
  text_in_english=[]
  for word in nltk_tokens:
    result = translate_text(word)
    if(result['detectedSourceLanguage']=='en'):
      b = TextBlob(word)
      corrected = b.correct()
      #print(corrected)
      word_in_eng = str(corrected)
      #print(sent_in_eng)
      #list_of_converted_words.append(word_in_eng)
      list_of_tagged_words.append(word_in_eng+"/E")
      list_of_english_words.append(word_in_eng)
    
    else:
      list_of_tagged_words.append(word+"/H")
      list_of_romanized_hindi_words.append(word)
      transliterated = transliterate(word, sanscript.ITRANS, sanscript.DEVANAGARI)
      #print(transliterated)
      list_of_devanagri_words.append(transliterated)
      result = translate_text(transliterated)
      list_of_translated_words.append(result['translatedText'])
    
    #print(result['translatedText'])
  translated_to_lower =  [x.lower() for x in list_of_translated_words]
  final_list= list_of_english_words + translated_to_lower

  #print(list_of_tagged_words)
  #print(list_of_english_words)  
  #print(list_of_romanized_hindi_words)
  #print(list_of_devanagri_words)
  #print(translated_to_lower)
  #print(final_list)
  listToStr = ' '.join(map(str, final_list))
  #print(listToStr)

  list_of_converted_sentences.append(listToStr) 
  list_of_all_tagged_words.append(list_of_tagged_words)
    
  score = sentiment_score_calc(final_list)

  score_list.append(score)
  #print(score_list)
```
